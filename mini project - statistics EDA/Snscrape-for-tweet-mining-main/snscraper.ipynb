# Mining Tweets on twitter using Snscrape

#### A simple step by step guide.


Make sure you have Snscrape installed on your machine and if it's not, quickly pip install snscrape. NB: You might want ti install the developer version on git. Once that is done, feel free to follow the steps in this notebook


```python
#import all the needed libraries

import snscrape
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Set the maximum number of variables to be used below.
maxTweets = 10

# Create an empty list to append tweet data to.
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list. The keyword in this code is for a project i worked on.

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('peter obi since:2022-03-04 until:2022-03-30-filter:links -filter:replies -filter:retweets').get_items()):
    if i>maxTweets:
        break
    tweets_list.append([tweet.date, tweet.id, tweet.content,tweet.sourceLabel, tweet.username, tweet.retweetCount, tweet.likeCount, tweet.quoteCount,
                   tweet.replyCount, tweet.user.followersCount])
#filter links and replies if you don't want to include retweets, replies and links to tweet.
```

```python
# Creating a dataframe from the tweets list above.

tweets_df = pd.DataFrame(tweets_list, columns=['date', 'id', 'content','Source', 'username', 
                                                 'retweetCount', 'likeCount', 'quoteCount',
                                                 'replyCount', 'followersCount'])

# Display first 5 entries from dataframe
tweets_df.head(5)
```

#### An helper function that you can re-use in case you want to run the search multiple times.

```python
def get_tweets(max_tweets, key_word):
    
    maxTweets = max_tweets

    tweets_list = []

    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(key_word).get_items()):
        if i>maxTweets:
            break
        tweets_list.append([tweet.date, tweet.id, tweet.content,tweet.sourceLabel, tweet.username, tweet.retweetCount, tweet.likeCount, tweet.quoteCount,
                   tweet.replyCount, tweet.user.followersCount])

    
    tweets_df = pd.DataFrame(tweets_list, columns=['date', 'id', 'content','Source', 'username', 
                                                 'retweetCount', 'likeCount', 'quoteCount',
                                                 'replyCount', 'followersCount'])
    
    return tweets_df.head(5)
```
